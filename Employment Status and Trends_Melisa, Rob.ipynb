# Employment Status and Trends

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
employment_data = pd.read_csv('CPSData.csv')
employment_data.head(8)
```

```python
# Employment Status by Age Group
age_data = employment_data.groupby(['Age','EmploymentStatus']).size().unstack().fillna(0)
```

```python
# Using Stacked bar chart show patterns of employment, unemployment, and participation across age groups
age_data.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.title('Employment Status and Trends Across Age Groups')
plt.xlabel('Age')
plt.ylabel('Number of Individuals')
plt.legend(title='Employment Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization/Employment Status by Age Group: This bar chart provides a clear visual representation of how employment status varies by age, allowing you to identify trends, disparities, and insights related to employment and participation in the labor market across different age demographics.
```

```python
#Employment Status by Gender
gender_data = employment_data.groupby(['EmploymentStatus', 'Married']).size().unstack().fillna(0)
```

```python
#Using Stacked bar chart show patterns of employment, unemployment, and participation across gender groups
gender_data.plot(kind='bar', stacked=True)
plt.title('Employment Status and Trends by Marrital Status')
plt.xlabel('Gender')
plt.ylabel('Number of Individuals')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
# Visualization/Employment Status by Marrital Status: The stacked bar chart provides a visual comparison of employment, unemployment, and participation rates across various marital status categories, such as single, married, divorced, and widowed. 
```

```python
#Employment Status by Education Level
education_level = employment_data.groupby(['Sex','Education']).size().unstack().fillna(0)
```

```python
#Using Stacked bar chart show patterns of employment, unemployment, and participation across gender groups
education_level.plot(kind='bar', stacked=True)
plt.title('Employment Status and Trends by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Individuals')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
# Visualization/Employment Status by Education Level: The chart allows for a visual comparison of employment, unemployment, and participation across various education levels (e.g., high school, bachelor's degree, master's degree)
```

```python
# Summary: The combined insights from these charts can inform policymakers and organizations about the current state of the labor market. Understanding how age, gender, and education intersect can guide the development of targeted policies and programs aimed at improving employment outcomes for specific demographics.
```
